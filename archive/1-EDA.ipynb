# Data preprocessing
Before the data can be used to perform any machine learning task we need to modify a bit its form. And for this we will describe our pipeline
1. (optional) *Map the Milestone* to usable labels (for binary classification we need to have only 0s and 1s)
2. **One-Hot encoding** of the hardware model (it is a categorical variable)
3. **Handling Missing Values** (It happens quite a lot that measurement are missing and we need a strategy to handle such cases).
4. **Scaling** the data (such that dimensions are comparable, some models won't work if this isn't done)

````python
y = df['milestone_name']
x = df[feature_cols]
````

## Getting binary labels
We will convert the milestone name because we want to interpret it as :
* if the milestone name is ```NaN``` then it means that the CPE is supposed to be ```healthy``` *(0)*
* otherwise it is considered as ```sick``` *(1)*

````python
y_binary = y.isnull().map(lambda x: 0 if x else 1)
````

##  One-hot encoding

````python
print('Different types :\t',set(x.dtypes.values))
print('Discrete features :\t',list(x.dtypes[x.dtypes == 'int64'].index))
````

X only contains two discrete variables:
* ```hardware_model```: this integer is simply an index referencing which model this corresponds to and therefore we do not wish to have a notion of ordering between models that might affect our model. 
* ```n_cpe_building```: it represents the number of CPE that are in the building, this will be scaled later on but we do not need to encode it, as explained above it carries a notion of ordering and therefore we do not mind

````python
x_encoded = pd.get_dummies(x, columns = ['hardware_model'],prefix =  "model")
````

## Handling missing values

We can see that NULL value are quite present in our dataset and we need A strategy to handle them. Many imputation techniques exists that often use to hide those missing values (replacing them by the mean of the feature for example or by a 0). Here we wish our missing values to be meaningful as they could really indicate a problem with this particular CPE, this scenario is often referred to as **Missing Not at Random (MNAR)**.

Nevertheless we have already added to each feature a corresponding ```MISS_``` feature that gives us an indication on the percentage of samples that are missing for the feature computation (because usually those features are aggregates of multiple values).

Therefore we wish to hide as much as possible the irregularity of such missing value and will use the mean.

````python
# if we want to replace them by zero
x_no_missing = x_encoded.fillna(0)

# but it might be more meaningful to use the median because we wish to erase as much as possible those missing values 
median_null_imputer = sklearn.preprocessing.Imputer(strategy='mean')
x_no_missing = median_null_imputer.fit_transform(x_encoded)
````

## Scaling the data
Now we can normalize our features. We won't justify the use for standardization as we will only be able to judge this step once we start applying some clustering algorithms. 

````python
x_scaled = pd.DataFrame(preprocessing.scale(x_no_missing), columns = x_encoded.columns)
x_scaled.head()
````

## Dimensionality reduction
Due to the high dimensionality of our dataset we wish to find the features that encapsulates the most information, which is why we use PCA: this method allows to detect columns that are highly correlated.

### Principal Component Analysis (PCA)

First we look at the repartition of explained variance ratio among the components, in order to have a visual interpretation of the PCA. We will see the cumulative variance retained depending on the number of components we chose.

````python
# We want to keep the features that allow us to explain 95% of the variance.
pca = PCA()
pca.fit(x_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
````

The curve isn't very steep which isn't quite a good news for us given that it forces us to use a lot of components in order to encapsulate a good part of the variance.

#### Choosing a PCA 
Now we need to chose the PCA dimensionality reduction that will yield the best results

````python
explained_variance = 0.95
pca = PCA(explained_variance)
pca.fit(x_scaled)
dimensions = len(pca.components_)

x_ready = pca.transform(x_scaled)
print("We retained {} dimension(s) to explain {}% of the variance".format(dimensions,100*explained_variance))
````

### Investigating the curse of dimensionality
We want to decide whether our clustering algorithm can work or not. The curse of dimensionality might prevent it from doing so, in order to determine this we can look at the distances between vectors and check whether they are meaningful or they all converge toward the same value. 

And to do that we also investigate multiple distance metrics.

````python
import scipy
dist = scipy.spatial.distance.pdist(x_ready)
np.save(DATA_FOLDER+'sample_dist.pk',dist)
````

### Balancing the two classes
One of the assumption of K-means is the fact that clusters should have approximately the same size. Therefore we try to subsample the 'believed healthy' class.

````python
# we split our x into two populations
vec_healthy = x_ready[y_binary == 0]
vec_sick = x_ready[y_binary == 1]
n_sick = len(vec_sick)

# we select as many random indices from vec_healthy as we have routers that are sick
sampled_indices = np.random.choice(len(vec_healthy), size=n_sick)

sample_healthy = vec_healthy[sampled_indices]
sample_healthy.shape
full_sample = np.vstack((sample_healthy,vec_sick))
sample_y = np.hstack((np.zeros(n_sick),np.ones(n_sick)))

K_means = sklearn.cluster.MiniBatchKMeans(n_clusters=2,batch_size=1000)
labels = K_means.fit_predict(full_sample)

get_single_silhouette(full_sample,labels,2)
````

Now that we have balanced classes we can look at the distribution of sick and healthy CPE inside each cluster.

````python
analyse_clustering(sample_y,labels)
````

From this we cannot really conclude on whether this clustering is meaningful. We can observe that:
* One of the cluster contains ~80% of **sick CPEs** 
* But the number of false negatives (not identified - in terms of clustering - sick CPEs)

this could mean either that our clustering isn't performing well or that our initial labels can't be trusted.
