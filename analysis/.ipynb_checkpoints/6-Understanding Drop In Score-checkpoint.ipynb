```python
# Some important imports
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
sns.set_context('notebook')

# Sklearn imports
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split

# Own Scripts import
from scripts.utils import *
from scripts.preprocessing import *
from scripts.plot import *

# get rid of warning due to deprecated modules in sklearn
import warnings
warnings.simplefilter('ignore')
```

```python
DATA_FOLDER = './Data'
s_day = '22_05'
DATA_PATH = DATA_FOLDER + '/sample_'+s_day+'.xlsx'
BACKUP_PATH = DATA_FOLDER+'/sample_'+s_day+'.pk'


extracted = import_sample(DATA_PATH,BACKUP_PATH)
dates = extracted['day_0'].values
x_extracted, y_extracted = get_ml_data(extracted)

y = convert_to_binary_labels(y_extracted)
x_encoded = encode_categorical(x_extracted)
x_df = remove_features(x_encoded,verbose = True)
x = x_df.values
```

We have seen that a cross validation on the the full dataset (without considering the necessity to have an empty intersection in terms of dates) was yielding an overly optimistic estimate of the performance. Our hypothesis is that the days of the week have some important influence on the constructed vectors.

The challenge as always is that we have high dimensional vectors which doesn't help the visualization. Therefore we will try to perform a PCA to look at the different days. 

# Weekday Distribution Influence
## Using PCA for dimension reduction

```python
day_of_week = x_extracted['weekday']
preprocessing_pipeline = make_pipeline(sklearn.preprocessing.Imputer(strategy='mean'),
                preprocessing.StandardScaler())
```

```python
X = preprocessing_pipeline.fit_transform(x)
x_2d = PCA(2).fit_transform(X)
```

```python
fig, axes = plt.subplots(nrows=7, sharex = True, sharey = True,figsize=(10, 5*7))
to_day_name = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
for d in range(0,7):
    vec = x_2d[day_of_week == d]
    axes[d].hexbin(vec[:,0],vec[:,1],bins='log',mincnt=2)
    axes[d].set_title('Day = {}'.format(to_day_name[d]))
```

## Using the 2 most important features
Rather than using the PCA we can look at the influence of the weekday based on the two most important features (according to the classifier)

```python
top_2_f = ['offline_pct','miss_snr_dn_6h']
```

```python
x_2D = preprocessing_pipeline.fit_transform(x_df[top_2_f].values)
fig, axes = plt.subplots(nrows=7, sharex = True, sharey = True,figsize=(10, 5*7))
to_day_name = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
for d in range(0,7):
    vec = x_2D[day_of_week == d]
    axes[d].hexbin(vec[:,0],vec[:,1],bins='log',mincnt=2)
    axes[d].set_title('Day = {}'.format(to_day_name[d]))
```

The top features being categorical we do not really see much on this last plot.

## Clustering

```python
x_b, y_b = get_balanced_classes(x,y)
X = preprocessing_pipeline.fit_transform(x_b)
K_means = sklearn.cluster.MiniBatchKMeans(n_clusters=2,batch_size=600)
labels = K_means.fit_predict(X)
```

```python
analyse_clustering(y_b,labels)
```

```python
df_b = pd.DataFrame(x_b,columns=x_df.columns)
df_b['cluster'] = labels
x_2d = PCA(2).fit_transform(X)
```

```python
df_b
```

```python
def dummy_2_wkday(x):
    if x['wk_0'] == 1:
        return 0
    elif x['wk_1'] == 1:
        return 1
    elif x['wk_2'] == 1:
        return 1
    elif x['wk_3'] == 1:
        return 3
    elif x['wk_4'] == 1:
        return 4
    elif x['wk_5'] == 1:
        return 5
    elif x['wk_6'] == 1:
        return 6
    else : return -1
    
def dummy_2_mdls(x):
    if(x['model_0'] == 1):
        return 0
    elif(x['model_1'] == 1):
        return 1
    elif(x['model_2'] == 1):
        return 2
    elif(x['model_3'] == 1):
        return 3
    elif(x['model_4'] == 1):
        return 4
    elif(x['model_5'] == 1):
        return 5
    elif(x['model_6'] == 1):
        return 6
```

```python
df_b['weekday'] = df_b.apply(dummy_2_wkday,axis=1)
df_b['models'] = df_b.apply(dummy_2_mdls,axis=1)
```

```python
c0 = df_b[df_b['cluster']==0]['weekday'].value_counts()
c1 = df_b[df_b['cluster']==1]['weekday'].value_counts()
weekday_prop = pd.DataFrame()
weekday_prop['cluster 0'] = round(100*c0/c0.sum(),2)
weekday_prop['cluster 1'] = round(100*c1/c1.sum(),2)
weekday_prop.sort_index()
```

```python
c0 = df_b[df_b['cluster']==0]['models'].value_counts()
c1 = df_b[df_b['cluster']==1]['models'].value_counts()
models_prop = pd.DataFrame()
models_prop['cluster 0'] = round(100*c0/c0.sum(),2)
models_prop['cluster 1'] = round(100*c1/c1.sum(),2)
models_prop.sort_index()
```
