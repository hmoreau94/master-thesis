````python
DATA_FOLDER = './Data'
s_day = '22_05'
DATA_PATH = DATA_FOLDER + '/sample_'+s_day+'.xlsx'
BACKUP_PATH = DATA_FOLDER+'/sample_'+s_day+'.pk'


extracted = import_sample(DATA_PATH,BACKUP_PATH)
dates = extracted['day_0'].values
x_extracted, y_extracted = get_ml_data(extracted)

y = convert_to_binary_labels(y_extracted)
x_encoded = encode_categorical(x_extracted)
x_df = remove_features(x_encoded,verbose = True)
x = x_df.values
````

We will now try to look at the feature importance in order to understand how does the classification works inside and to be able to go to HFC support with these information and perform a quick reality check. 

Many classifiers at this point have the same degree of performance, nevertheless we will use ```Gradient Boosting``` in order to perform this exploration as it has a built in function to perform such task.

````python
params = [int(x) for x in np.linspace(20,50,5)]
pipeline = lambda param: make_pipeline(sklearn.preprocessing.Imputer(strategy='mean'),preprocessing.StandardScaler()
                                       ,sklearn.ensemble.GradientBoostingClassifier(n_estimators=param))
precision_recall_curves(params,pipeline,x,y,'GB + Scaling',[0.8,0.9],dates)
````

````python
clf = make_pipeline(sklearn.preprocessing.Imputer(strategy='mean'),
                    preprocessing.StandardScaler(), 
                    sklearn.ensemble.GradientBoostingClassifier(n_estimators=175))
clf.fit(x,y)
````

````python
features = list(x_df.columns)
feature_analysis = pd.DataFrame()
feature_analysis['feature'] = features
feature_analysis['importance'] = clf.steps[2][1].feature_importances_
feature_analysis = feature_analysis.sort_values(by='importance',ascending=False)
````

# Investigating the distribution of the top 4 features

````python
n_top = 4
to_analyse = feature_analysis['feature'].head(n_top).values
df = pd.DataFrame(x_df.filter(items=list(to_analyse)))
df['sick'] = y
df
````

````python
def compare_distrib(var_to_explore, n_bins):
    binned_name = var_to_explore + '_binned'
    min_ = df[var_to_explore].min()
    max_ = df[var_to_explore].max()
    df[str(binned_name)] = pd.cut(
        df[var_to_explore], n_bins, include_lowest=True)

    cnt = pd.DataFrame()
    cnt['sick'] = df[df['sick'] == 1][binned_name].value_counts()
    cnt['healthy'] = df[df['sick'] == 0][binned_name].value_counts()
    cnt = cnt.apply(lambda x: round(100*x/x.sum(), 3)).sort_index()

    plt.style.use('ggplot')
    ind = np.array([i for i, _ in enumerate(cnt.index)])
    width = 0.35
    plt.bar(ind, cnt['sick'], width, label='sick')
    plt.bar(ind+width, cnt['healthy'], width, label='healthy')
    plt.xticks(ind + width / 2, cnt.index)
    plt.legend(loc='best')
    plt.xticks(rotation=90)
    plt.title(var_to_explore)
    plt.show()
````

````python
compare_distrib('offline_pct',10)
````

````python
compare_distrib('miss_snr_dn_6h',10)
````

````python
compare_distrib('cmts_utilization_dn_2d',10)
````

````python
compare_distrib('cmts_rx_dn_3d',15)
````
